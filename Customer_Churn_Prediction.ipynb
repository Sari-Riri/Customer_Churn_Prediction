# Prediction using Decision Tree

```
# define a function that takes in class (target variable vector) and finds the entropy of that class.
# the fraction is â€˜piâ€™, it is the proportion of a number of elements in that split group to the number of elements in the group before splitting(parent group).
def find_entropy(train):
    entropy = 0
    for value in ('Yes', 'No'):
        fraction = train['Churn'].value_counts()[value] / len(train['Churn'])
        entropy += -fraction * np.log2(fraction)
    return entropy
```

```
train['Churn'].value_counts()['No']
```

```
# find the Entropy and then Information Gain for splitting the data set

find_entropy(train)
```

```
def find_entropy_attribute(train, attribute):
    entropy_sum = 0
    for variable in train[attribute].unique():
        entropy = 0
        for target_variable in ('Yes', 'No'):
            num = len(train[attribute][train[attribute] == variable][train['Churn'] == target_variable])
            den = len(train[attribute][train[attribute] == variable])
            fraction = num / (den + eps)
            entropy += -fraction * np.log2(fraction + eps)
        entropy_weights = den / len(train)
        entropy_sum += -entropy_weights * entropy
    return abs(entropy_sum)
```

```
def build_tree(train):
    IG = []
    for key in train.keys()[:-1]:
        IG.append(find_entropy(train) - find_entropy_attribute(train, key))
    node = train.keys()[:-1][np.argmax(IG)]  # Get attribute with maximum information gain

    tree = {}
    tree[node] = {}

    for value in np.unique(train[node]):
        sub_table = train[train[node] == value].reset_index(drop=True)
        cls, counts = np.unique(sub_table['Churn'], return_counts=True)

        if len(counts) == 1:  # Checking purity of subset
            tree[node][value] = cls[0]
        else:
            tree[node][value] = build_tree(sub_table)  # Calling the function recursively

    return tree
```

```
# Now we call the buildTree function and print the tree we built.
tree = build_tree(train)

import pprint
pprint.pprint(tree)
```

```
# We can write an algorithm to predict using this tree structure.
def predict(inst, tree):
    for nodes in tree.keys():
        value = inst[nodes]
        tree = tree[nodes][value]
        prediction = 0

        if type(tree) is dict:
            prediction = predict(inst, tree)
        else:
            prediction = tree
            break

    return prediction
```
